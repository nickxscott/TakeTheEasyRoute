```python
import pandas as pd
import numpy as np
import geopandas as gpd
import gpxpy

import shapely
import contextily as cx
from IPython.core.pylabtools import figsize
import matplotlib.pyplot as plt
import seaborn as sns

from typing import Dict, Union
from datetime import datetime
```

First, I need to parse the .gpx files I've downloaded from Strava and load them into a dataframe. The code to do just that has been graciously privided  by Alan Bunbury, and can be found here: https://github.com/bunburya/fitness_tracker_data_parsing/blob/main/parse_gpx.py . 

```python
# The XML namespaces used by the GPX file for extensions, used when parsing the extensions
NAMESPACES = {'garmin_tpe': 'http://www.garmin.com/xmlschemas/TrackPointExtension/v1'}

# The names of the columns we will use in our DataFrame
COLUMN_NAMES = ['latitude', 'longitude', 'elevation', 'time', 'heart_rate', 'cadence']

def get_gpx_point_data(point: gpxpy.gpx.GPXTrackPoint) -> Dict[str, Union[float, datetime, int]]:
        """Return a tuple containing some key data about `point`."""
        
        data = {
            'latitude': point.latitude,
            'longitude': point.longitude,
            'elevation': point.elevation,
            'time': point.time
        }
    
        # Parse extensions for heart rate and cadence data, if available
        elem = point.extensions[0]  # Assuming we know there is only one extension
        try:
            data['heart_rate'] = int(elem.find('garmin_tpe:hr', NAMESPACES).text)
        except AttributeError:
            # "text" attribute not found, so data not available
            pass
            
        try:
            data['cadence'] = int(elem.find('garmin_tpe:cad', NAMESPACES).text)
        except AttributeError:
            pass

        return data
    
def get_dataframe_from_gpx(fname: str) -> pd.DataFrame:
    """Takes the path to a GPX file (as a string) and returns a Pandas
    DataFrame.
    """
    with open(fname) as f:
        gpx = gpxpy.parse(f)
    segment = gpx.tracks[0].segments[0]  # Assuming we know that there is only one track and one segment
    data = [get_gpx_point_data(point) for point in segment.points]
    return pd.DataFrame(data, columns=COLUMN_NAMES)
```

```python
LESloop_df = get_dataframe_from_gpx('../data/LES_loop.gpx')
```

```python
LESloop_df.head()
```

Now I need this in the format of a GeoDataFrame in order to generate maps

```python
LESloop_gdf = gpd.GeoDataFrame(LESloop_df, geometry=gpd.points_from_xy(LESloop_df.longitude, LESloop_df.latitude))
```

```python
LESloop_gdf.head()
```

This looks good, so I'm going to export this as a .csv to use in a later notebook

```python
LESloop_gdf.to_csv('../data/LESloop_gdf.csv')
```

Now to plot the points on a map

```python
fig, ax = plt.subplots(figsize=(12,12))

ax=LESloop_gdf.plot(ax=ax, alpha=0.5, markersize=15, color='navy');

basemap=cx.providers.CartoDB.Voyager

cx.add_basemap(ax, crs=4326, source=basemap)

plt.title("Map of LES Loop");
```

Now that I have the route on a map, I am curious to see what some of the heartrate data looks like

```python
LES_hr = LESloop_gdf['heart_rate'].to_numpy()
```

```python
fig, ax = plt.subplots(figsize=(10,8))

sns.histplot(LES_hr, stat='percent', binwidth=2, kde=True, ax=ax)

ax.lines[0].set_color('crimson')

plt.title("LES Loop Heart Rate Distribution");
```
